# Criando uma curva de luz

Vamos criar uma instância STP com duração de 90 dias e uma cadência de 0.020833 (são os valores padrões da missão Kepler). 

```python
from simtransp import STP
stp = STP(tmin = 0, tmax = 90, cadencia = 0.020833, ai = 90)
```

A instância `STP` corresponde à curva de luz de uma estrela de disco uniforme. Vamos plotar para ver isso:

```python
stp.plot();
```

# Adicionando spots e fáculas

Precisamos chamar o `add_spots` e definir o período de rotação da estrela

```python
stp.add_spots(Peq = 28)
stp.plot(titulo = 'Curva de luz com spots e fáculas');
```

# Adicionando ruído branco

Definindo um desvio padrão e chamando o atributo `add_noise`, podemos plotar o gráfico atualizado:

```python
std = 1000e-6
stp.add_noise(std = std)
stp.plot(titulo = 'Curva de luz spots + ruído');
```

# Adicionando trânsito planetário

Precisamos especificar:
- raio da estrela `Re` em raios solares
- o raio do planeta `Rp` em razão do raio solar
- Período orbital do planeta

```python
stp.add_transito(Re = 1, Rp = 0.15, Porb = 10)
stp.plot(titulo = 'Curva de luz com spots + ruído + trânsito planetário');
```
